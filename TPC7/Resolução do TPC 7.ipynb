```python
#Função derivada:
# poinomios com estrutura ax^b

def derivada(pol):
    lista_termos = []
    for termo in pol:
        a = termo[0] * termo[1]
        b = termo[1] - 1
        lista_termos.append((a,b))
    return lista_termos

p1 = [(2, 3), (7, 4)]
print (derivada(p1))
```

```python
#Soma de polinómios:

#USAR DICIONARIOS

def soma_pol(pol1, pol2):
    novo_pol = []
    for termo1, termo2 in pol1, pol2:
        if termo1[1] == termo2[1]:
            a = termo1[0] + termo2[0]
            b = termo1[1]
            novo_pol.append((a,b))  
        elif termo1[1] != termo2[1]:
            c = termo1[0] 
            d = termo1[1] 
            novo_pol.append((c,d))
        elif termo2[1] != termo1[1]:
            e = termo2[0] 
            f = termo2[1] 
            novo_pol.append((e,f))
    return novo_pol
p1 = [(2, 3), (7, 4)]
p2 = [(2,3), (4,4)]
print(soma_pol(p1, p2))
```

```python
#Gráfico:

import matplotlib.pyplot as plt

x = []
y = []
for termo in p1:
    x.append(termo[0])
    y.append(termo[1])

plt.plot(x, y)
plt.xlabel("Coeficiente")
plt.ylabel("Expoente")
plt.title("Polinómios")
plt.show()
```
